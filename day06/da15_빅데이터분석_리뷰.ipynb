#### 판다스 데이터타입
|dtype|설명|비고
|------|---|------|
|**float64**|실수|기본값
|float32,16|실수(작은크기)|옵션(float64에서 변경가능)|
|**int64**|정수|기본값|
|int32,16,8|정수(작은크기)|옵션|
|datetime64|날짜| |
|**object**|문자열 또는 복합형| |
|bool|참거짓| |
|category|카테고리| |

```python
## 판다스 카테고리
df = pd.DataFrame({'id':[1,2,3,4,5,6],
                   'raw_grade':['a','a','b','b','c','d']
                   })
```

```python
df
```

```python
df.info()
```

```python
## 카테고리 변환
df['grade'] = df['raw_grade'].astype('category')
```

```python
df
```

```python
df.info()
```

```python
df['grade']
```

```python
df['grade']
```

```python
df.sort_values(by='grade')
```

```python
df.groupby('grade').size()
```

```python
# a,b,c,d를 다른 그룹핑 이름으로 변경 가능
df['grade'].cat.rename_categories(['Very good', 'good', 'bad', 'Very bad'])
```

```python
# 복사본 만들기
dfCopy = dfTitanic.copy()
```

```python
dfCopy[['survived', 'age', 'fare']].tail()
```

```python
# 필요 row만 가져올 때 loc속성
dfCopy.loc[0:5]
```

```python
# 필요 row만 가져올 때 loc속성 연속된 값을 가져올 때(파이썬 리스트 slice와 유사)
# 인덱스가 번호의 문자로 인덱스가 되어있을때 row를 가져오는 방법
dfCopy.loc[0:1]
```

```python
# 필요 row만 가져올 때, 필요한 row만 따로 가져올 때
dfCopy.loc[[0,4,23]]
```

```python
# 필요row만 가져올 때
dfCopy.iloc[0]
```

```python
# 필요한 row를 가져올때 행번호로 선택
#번호를 하나만 쓰면 그 인덱스 해당한 row 값만 iloc은 행과 열의 인덱스로만 해당
dfCopy.iloc[:, 3]
```

```python
# 결측치 검색, 전체합계 sum()
dfCopy.isnull().sum()
```

```python
# 결측치 검색 isna()
dfCopy.isna().sum()
```

```python
# 전체 결측치 합
dfCopy.isna().sum().sum()
```

```python
# 결측데이터만 필터링
condition = dfCopy['age'].isnull()
dfCopy[condition]
```

```python
# 두번째 방법
dfCopy.loc[condition]
```
